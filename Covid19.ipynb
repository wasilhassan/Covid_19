```python
import pandas as pd
import plotly.graph_objects as go


# Filter the data for state = "WA"
wa_data = df[df['state'] == 'WA']

# Extract the year from the 'submission_date' column
wa_data['year'] = pd.to_datetime(wa_data['submission_date']).dt.year.astype(int)

# Group the filtered data by year and calculate the total cases for each year
wa_total_cases = wa_data.groupby('year')['tot_cases'].max().reset_index()

# Create the bar graph using Plotly
fig = go.Figure(data=[go.Bar(x=wa_total_cases['year'], y=wa_total_cases['tot_cases'])])

# Add hover information
fig.update_traces(hovertemplate='Year: %{x}<br>Total Cases: %{y}')

# Set the labels and title for the graph
fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Total Cases',
    title='Total COVID-19 Cases in Washington (WA) by Year'
)

# Remove the decimal part from the x-axis values
fig.update_xaxes(tickmode='linear', tickformat='d')

# Show the interactive graph
fig.show()
```

```python
import pandas as pd

# Filter the data for state = "WA" and state = "TX"
wa_data = df[df['state'] == 'WA']
tx_data = df[df['state'] == 'TX']

# Extract the 'tot_cases' column for WA and TX
wa_total_cases = wa_data['new_case']
tx_total_cases = tx_data['new_case']

# Calculate descriptive statistics for WA
wa_total_cases_count = wa_total_cases.count()
wa_total_cases_mean = wa_total_cases.mean()
wa_total_cases_median = wa_total_cases.median()
wa_total_cases_std = wa_total_cases.std()
wa_total_cases_min = wa_total_cases.min()
wa_total_cases_max = wa_total_cases.max()

# Calculate descriptive statistics for TX
tx_total_cases_count = tx_total_cases.count()
tx_total_cases_mean = tx_total_cases.mean()
tx_total_cases_median = tx_total_cases.median()
tx_total_cases_std = tx_total_cases.std()
tx_total_cases_min = tx_total_cases.min()
tx_total_cases_max = tx_total_cases.max()

# Create a DataFrame to display the statistics
statistics = pd.DataFrame({
    'Statistic': ['Count', 'Mean', 'Median', 'Standard Deviation', 'Minimum', 'Maximum'],
    'Washington (WA)': [wa_total_cases_count, wa_total_cases_mean, wa_total_cases_median, wa_total_cases_std, wa_total_cases_min, wa_total_cases_max],
    'Texas (TX)': [tx_total_cases_count, tx_total_cases_mean, tx_total_cases_median, tx_total_cases_std, tx_total_cases_min, tx_total_cases_max]
})

# Format the numbers with comma separators and 2 decimal places
statistics['Washington (WA)'] = statistics['Washington (WA)'].apply(lambda x: "{:,.2f}".format(x))
statistics['Texas (TX)'] = statistics['Texas (TX)'].apply(lambda x: "{:,.2f}".format(x))

# Display the statistics
print("Descriptive Statistics for New Cases:")
print("--------------------------------------")
print(statistics)
```

```python
import pandas as pd
import plotly.express as px

# Filter the data for state = "WA" and state = "TX"
wa_data = df[df['state'] == 'WA']
tx_data = df[df['state'] == 'TX']

# Group the data by submission_date and calculate the total deaths
wa_total_deaths = wa_data.groupby('submission_date')['new_death'].sum().reset_index()
tx_total_deaths = tx_data.groupby('submission_date')['new_death'].sum().reset_index()

# Concatenate the dataframes with a new 'State' column
wa_total_deaths['State'] = 'WA'
tx_total_deaths['State'] = 'TX'
combined_data = pd.concat([wa_total_deaths, tx_total_deaths])

# Create the line graph
fig = px.line(
    combined_data,
    x='submission_date',
    y='new_death',
    color='State',
    labels={'submission_date': 'Date', 'new_death': 'Total Deaths'},
    title='Death per day Comparison: WA vs TX'
)

# Display the graph
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Convert 'submission_date' to datetime type
df['submission_date'] = pd.to_datetime(df['submission_date'])

# Filter the data for state = "WA" and state = "TX"
wa_data = df[df['state'] == 'WA']
tx_data = df[df['state'] == 'TX']

# Group the data by year and calculate the total deaths
wa_total_deaths = wa_data.groupby(wa_data['submission_date'].dt.year)['new_death'].sum().reset_index()
tx_total_deaths = tx_data.groupby(tx_data['submission_date'].dt.year)['new_death'].sum().reset_index()

# Reset the index and rename the columns
wa_total_deaths = wa_total_deaths.reset_index().rename(columns={'submission_date': 'Year', 'new_death': 'WA Total Deaths'})
tx_total_deaths = tx_total_deaths.reset_index().rename(columns={'submission_date': 'Year', 'new_death': 'TX Total Deaths'})

# Create the line graph
fig = px.line(
    pd.concat([wa_total_deaths, tx_total_deaths]),
    x='Year',
    y=['WA Total Deaths', 'TX Total Deaths'],
    labels={'Year': 'Year', 'value': 'Total Deaths'},
    title='Total Deaths Comparison: WA vs TX'
)

# Display the graph
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.colors as colors
import random


# Select five random states from the DataFrame
random_states = random.sample(list(df['state'].unique()), 5)

# Filter the data for the selected states
filtered_data = df[df['state'].isin(random_states)]

# Group the filtered data by state and calculate the total deaths
total_deaths = filtered_data.groupby('state')['new_death'].sum().reset_index()

# Sort the total deaths in descending order
total_deaths = total_deaths.sort_values('new_death', ascending=False)

# Create a color palette with unique colors for each state
color_palette = random.sample(colors.qualitative.Plotly, len(total_deaths))

# Format the total deaths as whole numbers with comma separators
total_deaths['formatted_deaths'] = total_deaths['new_death'].apply(lambda x: f'{int(x):,}')

# Create the funnel graph
fig = go.Figure(go.Funnel(
    y=total_deaths['state'],
    x=total_deaths['new_death'],
    text=total_deaths['formatted_deaths'],
    textposition="inside",
    orientation="h",
    marker=dict(color=color_palette)
))

fig.update_layout(
    title='Total Deaths Funnel Graph',
    xaxis_title='Total Deaths',
    yaxis_title='State'
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Choose a state for the pie chart
state = 'WA'  # Change this to the desired state

# Filter the data for the selected state
state_data = df[df['state'] == state]

# Extract the year from the submission_date column
state_data['year'] = pd.to_datetime(state_data['submission_date']).dt.year

# Calculate the total cases for each year in the selected state
total_cases_by_year = state_data.groupby('year')['new_death'].sum()

# Create the pie chart
fig = px.pie(
    names=total_cases_by_year.index,
    values=total_cases_by_year.values,
    title=f'Total COVID-19 Cases by Year in {state}'
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(showlegend=False)
fig.show()
```
